```python
%sql
CREATE OR REPLACE TABLE gold_daily_sales_fact
USING DELTA
CLUSTER BY (order_date, product_category, customer_id)
AS

SELECT
    t.order_id,
    t.customer_id,
    c.customer_key,
    t.item_id,
    p.product_name,
    p.category AS product_category,
    t.quantity,
    t.price,
    (t.quantity * t.price) AS total_amount,
    t.order_timestamp,
    DATE(t.order_timestamp) AS order_date,
    c.region
FROM silver_transactions t
JOIN silver_products p
ON t.item_id = p.item_id
JOIN silver_customers_scd2 c
ON t.customer_id = c.customer_id
-- AND t.order_timestamp >= c.start_date
AND (
    c.end_date IS NULL
    OR t.order_timestamp < c.end_date
);
```
